```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder='Desktop/crem/raw data'
os.listdir(folder)
```

```python
df=pd.read_csv(os.path.join(folder,'renewable_power_plants_CH.csv'))
df
df['commissioning_year']=df['commissioning_date'].astype(str).str[:4]
```

```python
%matplotlib notebook
plt.figure('Electrical capacity per year',figsize=(8,5))
plt.plot(df.groupby(['commissioning_year'])['electrical_capacity'].sum(),'-o',color='black',alpha=0.4)
plt.grid(color='xkcd:silver')
plt.xlabel('ANNÉE')
plt.ylabel('PUISSANCE')
```
