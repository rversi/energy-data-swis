```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'technologie': ['Chauffages à bois automatiques > 70 kW, avec mesures compl. relatives à l’épuration des fumées', 'Réseaux de chaleur à distance au bois', 'Mesures spéciales', 'Exploitation des rejets thermiques', 'Chauffages à bois automatiques > 70 kW, sans mesures compl. relatives à l’épuration des fumées', 'Chauffages à bois automatiques < 70kW', 'Nouveaux bâtiments MINERGIE', 'Capteurs solaires', 'Rénovation MINERGIE', 'Enveloppe du bâtiment, aération douce', 'Pompes à chaleur', 'Nouveaux bâtiments MINERGIE-P', 'Rénovation du système', 'Chauffage à bûches', 'Installations photovoltaïques', 'Rénovation CECB: efficacité énergétique globale B / efficacité enveloppe minimum C', 'Nouveaux bâtiments, systèmes', 'Rénovation MINERGIE-P', 'Rénovation installations MINERGIE', 'Rénovation CECB: efficacité énergétique globale A/ efficacité de l\'enveloppe minimum C'],
    'TWh économisés': [16.5, 14, 12.25, 11.5, 10.25, 8.5, 8.5, 6, 5.5, 5.25, 4, 3.5, 3, 3, 2, 1, 0.5, 0.25, 0.04, 0.03],
    'mio t. CO2 économisés': [3, 2.4, 2.6, 2.2, 2.1, 1.15, 1.25, 1.05, 1, 1, 1.05, 0.5, 0.55, 0.5, 0, 0.1, 0.05, 0.04, 0.01, 0.01],
    'sommes investies': [64, 53.5, 131.9, 48.9, 34.1, 41.5, 76.7, 178, 67.1, 104.5, 84.3, 64.9, 33.3, 22.6, 81.8, 13.1, 2.7, 4.3, 0.5, 2]
}

df = pd.DataFrame(data)
df['kWh/CHF_subv'] = 1e9*df['TWh économisés'] / (1e6*df['sommes investies'])
df['kg CO2/CHF_subv'] = 1e3*df['mio t. CO2 économisés'] / df['sommes investies']
```

```python
df['sommes investies 2016'] = [5, 4.6, 16.8, 5.3, 0.5, 1.8, 1.7, 7.3, 0.6, 2.8, 10.5, 11.1, 1.7, 0.6, 0.2, 1, 0, 0.6, 0, 0.4]
df['investissments générés 2016'] = [35, 17, 34, 11, 3, 7, 11, 46, 20, 12, 25, 60, 10, 2, 5, 1, 0, 3, 0, 1]
df['taux subventionnement 2016'] = 100*df['sommes investies 2016'] / df['investissments générés 2016']
df
```

```python
%matplotlib notebook

fig, ax1 = plt.subplots(figsize=(9,14))
ax1.bar(range(len(df['technologie'])), df['kWh/CHF_subv'], color='red', label='kWh/CHF', alpha=1, width=-0.3, align='edge')
ax1.set_xticks(range(len(df['technologie'])))
ax1.set_xticklabels(df['technologie'], rotation='vertical')
#ax1.legend(loc='best')

ax2 = ax1.twinx()
ax2.bar(range(len(df['technologie'])), df['kg CO2/CHF_subv'], color='black', label='kg CO2/CHF', alpha=1, width=0.3, align='edge')
#ax2.legend(loc='best')

ax3 = ax1.twinx()
ax3.bar(range(len(df['technologie'])), df['sommes investies'], color='blue', label='mio CHF', alpha=0.3, width=0.8)
ax3.spines['right'].set_position(('outward', 60))
ax3.legend(loc='best')

ax1.tick_params('y', colors='red')
ax2.tick_params('y', colors='black')
ax3.tick_params('y', colors='b')

plt.tight_layout()
```

```python
plt.figure('Taux de subventionnement', figsize=(9,12))
plt.bar(range(len(df['technologie'])), 100*df['sommes investies 2016']/(df['sommes investies 2016']+df['investissments générés 2016']), color='darkred', label='mio CHF', alpha=0.8, width=0.5)
plt.bar(range(len(df['technologie'])), 100*df['investissments générés 2016']/(df['sommes investies 2016']+df['investissments générés 2016']), color='darkred', label='kWh/CHF', alpha=0.2, width=0.5, bottom=100*df['sommes investies 2016']/(df['sommes investies 2016']+df['investissments générés 2016']))
plt.xticks(range(len(df['technologie'])), df['technologie'], rotation=90)
plt.yticks(np.arange(0, 61, 10), rotation=90)
plt.tight_layout()
plt.axis([-1,len(df['technologie']),0, 60])
plt.grid(color='xkcd:silver', alpha=0.3)
#plt.savefig('Taux de subv par techno')
```

```python
plt.figure('analyse 2', figsize=(9,9))
plt.scatter(df['kg CO2/CHF_subv'], df['kWh/CHF_subv'], s=15*df['sommes investies'], c=df.index, cmap='tab20b', alpha=0.7)
plt.xlabel('kg CO2 / CHF')
plt.ylabel('kWh / CHF')
plt.grid(color='xkcd:silver', alpha=0.4)
plt.savefig('analyse2_donnee_PB')
```
