```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data_folder='Desktop/CREM/Raw Data/OFS'
data_folder='Desktop/CREM/Data'
os.listdir(raw_data_folder)
```

```python
os.listdir(data_folder)

def changeinzero(x):
    if type(x)is str:
        return 0
    else:
        return x
```

```python
demographic_data_df=pd.read_excel(os.path.join(raw_data_folder,'Données démographie suisse.xlsx'),skiprows=2,skip_footer=42)
demographic_data_df.drop(demographic_data_df.index[1:4],inplace=True)
demographic_data_df=demographic_data_df.T
head_row=['Population résidente (en milliers)','Étrangers (en milliers)','Régions urbaines (en milliers)','Régions rurales (en milliers)']
demographic_data_df.set_index([0],inplace=True)
demographic_data_df.drop(demographic_data_df.index[0],inplace=True)
demographic_data_df.columns=head_row
demographic_data_df['Régions urbaines (en milliers)']=demographic_data_df['Régions urbaines (en milliers)'].map(changeinzero)
demographic_data_df['Régions rurales (en milliers)']=demographic_data_df['Régions rurales (en milliers)'].map(changeinzero)
demographic_data_df.to_csv(os.path.join(data_folder,'National demographic data'))
demographic_data_df.head()
```

```python
data_dict={}
year_numb=2016
for x in range (6):
    year_numb=str(year_numb)
    canton_demographic_data_df=pd.read_excel(os.path.join(raw_data_folder,'Démographie par canton.xlsx'),sheetname=x,skiprows=5,skip_footer=9)
    canton_demographic_data_df.drop(canton_demographic_data_df.columns[2:],axis=1,inplace=True)
    canton_demographic_data_df.columns=['Cantons','Population']
    canton_demographic_data_df.loc[0,'Cantons']='Suisse' #modifie la valeur d'une case
    canton_demographic_data_df.set_index('Cantons',inplace=True)
    canton_demographic_data_df['%']=100*canton_demographic_data_df['Population']/canton_demographic_data_df.loc['Suisse','Population']
    canton_demographic_data_df=canton_demographic_data_df.T
    split_index=[[year_numb,year_numb],['Population','Part en %']]
    year_numb=int(year_numb)
    split_index=pd.MultiIndex.from_arrays(split_index)
    split_index
    canton_demographic_data_df.set_index(split_index,inplace=True)
    data_dict[2016-x]=canton_demographic_data_df
    x+=1
    year_numb=int(year_numb)
    year_numb=year_numb-1
```
