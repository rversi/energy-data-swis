```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
df=pd.read_csv(os.path.join('../PRE_PROCESS_DATA/Données nbre bâtiments par agents énergétiques'))
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_suisse2015=df[(df['Zone géographique']=='Suisse')&(df['Année']==2015)]
df_valais2015=df[(df['Zone géographique']=='Valais / Wallis')&(df['Année']==2015)]
df_geneve2015=df[(df['Zone géographique']=='Genève')&(df['Année']==2015)]
df_vaud2015=df[(df['Zone géographique']=='Vaud')&(df['Année']==2015)]
df_zurich2015=df[(df['Zone géographique']=='Zürich')&(df['Année']==2015)]
df.head()
```

```python
fracs=df_suisse2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum()/df_suisse2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum().sum()
labels=['Autres','Bois','Capteurs solaires','CAD','Charbon','Électricité','Gaz','Mazout','PAC']
colors=['red','maroon','green','moccasin','black','yellow','silver','dimgray','gold']

fracs2=df_valais2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum()/df_valais2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum().sum()
fracs3=df_geneve2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum()/df_geneve2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum().sum()
fracs4=df_vaud2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum()/df_vaud2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum().sum()
fracs5=df_zurich2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum()/df_zurich2015.groupby(['Agent énergétique chauffage'])['Nombre de bâtiments'].sum().sum()

%matplotlib notebook
plt.figure('SUISSE = Part (nbre) des agents énergétiques pour le chauffage en 2015',figsize=(6,6))
plt.pie(fracs,labels=labels,startangle=90,colors=colors)
plt.axis('equal')

plt.figure('GENEVE = Part (nbre) des agents énergétiques pour le chauffage en 2015',figsize=(8,8))
plt.pie(fracs3,labels=labels,startangle=90,colors=colors,autopct='%.1f%%')

plt.figure('VALAIS = Part (nbre) des agents énergétiques pour le chauffage en 2015',figsize=(8,8))
plt.pie(fracs2,labels=labels,startangle=90,colors=colors,autopct='%.1f%%')

plt.figure('VAUD = Part (nbre) des agents énergétiques pour le chauffage en 2015',figsize=(8,8))
plt.pie(fracs4,labels=labels,startangle=90,colors=colors,autopct='%.1f%%')

plt.figure('ZURICH = Part (nbre) des agents énergétiques pour le chauffage en 2015',figsize=(8,8))
plt.pie(fracs5,labels=labels,startangle=90,colors=colors,autopct='%.1f%%')
```

```python
tot=df[(df['Zone géographique']!='Suisse')&(df['Année']==2015)].groupby('Zone géographique').sum()['Nombre de bâtiments']
x=[]
for i in df['Époque de construction'].unique():
    dff = df[(df['Époque de construction']==i)&(df['Zone géographique']!='Suisse')&(df['Année']==2015)].groupby(['Zone géographique'])['Nombre de bâtiments'].sum()
    x[i] = {sum(Idc[Idc['ÉpoqueConstruction']==i]['IDC moyen']):dff/tot}

x
```
