```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder='desktop/crem'
```

```python
df=pd.read_excel(os.path.join(folder,'Nbre bâtiment par agent énergétique.xlsx'))
df.reset_index(inplace=True)
df.fillna(method='ffill',inplace=True)
df.rename(columns={'level_0':'ID Canton','level_1':'Zone géographique','level_2':"Type d'habitation",'level_3':'Agent énergétique chauffage','level_4':'Agent énergétique ECS'},inplace=True)
```

```python
df1=df.iloc[0:,:5]
df1.drop([0,1],inplace=True)
df2=df1.copy()
for i in range(62): 
    df2=pd.concat([df2,df1])
```

```python
df2=df2.sort_index()
```

```python
df3=df.iloc[0:2,:]
df3=df3.drop(['ID Canton','Zone géographique',"Type d'habitation",'Agent énergétique chauffage','Agent énergétique ECS'],axis=1)
df3=df3.T
df4=df3.copy()
```

```python
for i in range(8747):
    
    df4=pd.concat([df4,df3])
```

```python
df4.reset_index(inplace=True)
df4=df4.drop(['index'],axis=1)
```

```python
df4.columns=['Année','Époque de construction']
```

```python
df2.reset_index(inplace=True)
df2=df2.drop(['index'],axis=1)
```

```python
DF=pd.merge(df2, df4, left_index=True, right_index=True)
```

```python
df5=df.iloc[2:,5:]
```

```python
df5=df5.T
```

```python
df6=df5.iloc[:,0]
for col in range (1,8748):
    Y=df5.iloc[:,col]
    df6=df6.append(Y)
```

```python
DF1=pd.DataFrame(df6)
DF1.reset_index(inplace=True)
DF1.drop(['index'],axis=1,inplace=True)
DF1.columns=['Nombre de bâtiments']
```

```python
DF_AGB=pd.merge(DF, DF1, left_index=True, right_index=True)
```
