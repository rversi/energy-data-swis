```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv(os.path.join('../PRE_PROCESS_DATA','Données nbre bâtiments par agents énergétiques'))
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
mazout=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Mazout (chauf.)')].groupby('Année').sum()
mazout['bâtiment par an']=mazout['Nombre de bâtiments']

for i in range(1,7):    
    mazout.iloc[i,1]-=mazout.iloc[i-1,2]
mazout=mazout.iloc[1:7,1]

gaz=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Gaz (chauf.)')].groupby('Année').sum()
gaz['bâtiment par an']=gaz['Nombre de bâtiments']

for i in range(1,7):    
    gaz.iloc[i,1]-=gaz.iloc[i-1,2]
gaz=gaz.iloc[1:7,1]

pac=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Pompe à chaleur (chauf.)')].groupby('Année').sum()
pac['bâtiment par an']=pac['Nombre de bâtiments']

for i in range(1,7):    
    pac.iloc[i,1]-=pac.iloc[i-1,2]
pac=pac.iloc[1:7,1]

elec=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Electricité (chauf.)')].groupby('Année').sum()
elec['bâtiment par an']=elec['Nombre de bâtiments']

for i in range(1,7):    
    elec.iloc[i,1]-=elec.iloc[i-1,2]
elec=elec.iloc[1:7,1]

bois=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Bois (chauf.)')].groupby('Année').sum()
bois['bâtiment par an']=bois['Nombre de bâtiments']

for i in range(1,7):    
    bois.iloc[i,1]-=bois.iloc[i-1,2]
bois=bois.iloc[1:7,1]

sol=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Capteur solaire (chauf.)')].groupby('Année').sum()
sol['bâtiment par an']=sol['Nombre de bâtiments']

for i in range(1,7):    
    sol.iloc[i,1]-=sol.iloc[i-1,2]
sol=sol.iloc[1:7,1]

cad=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Chaleur à distance (chauf.)')].groupby('Année').sum()
cad['bâtiment par an']=cad['Nombre de bâtiments']

for i in range(1,7):    
    cad.iloc[i,1]-=cad.iloc[i-1,2]
cad=cad.iloc[1:7,1]

aut=ep2015[(ep2015['Zone géographique']=='Suisse')&(ep2015['Agent énergétique chauffage']=='Autre agent énergétique (chauf.)')].groupby('Année').sum()
aut['bâtiment par an']=aut['Nombre de bâtiments']

for i in range(1,7):    
    aut.iloc[i,1]-=aut.iloc[i-1,2]
aut=aut.iloc[1:7,1]

total=ep2015[ep2015['Zone géographique']=='Suisse'].groupby(['Année']).sum()

total['Total']=total['Nombre de bâtiments']
for i in range(1,7):    
    total.iloc[i,1]-=total.iloc[i-1,2]
total=total.iloc[1:7,1]

plt.figure('Évolution de la part des nouveaux agents énergétiques (chauf.) installés en Suisse',figsize=(8,6))

width=0.35
N=np.arange(6)

plt.bar(N,100*mazout/total,width,color='grey')
plt.bar(N,100*gaz/total,width,bottom=100*mazout/total,color='silver')
plt.bar(N,100*pac/total,width,bottom=100*mazout/total+100*gaz/total)
plt.bar(N,100*elec/total,width,bottom=100*mazout/total+100*gaz/total+100*pac/total,color='gold')
plt.bar(N,100*bois/total,width,bottom=100*mazout/total+100*gaz/total+100*pac/total+100*elec/total,color='maroon')
plt.bar(N,100*sol/total,width,bottom=100*mazout/total+100*gaz/total+100*pac/total+100*elec/total+100*bois/total,color='green')
plt.bar(N,100*cad/total,width,bottom=100*mazout/total+100*gaz/total+100*pac/total+100*elec/total+100*bois/total+100*sol/total)
plt.bar(N,100*aut/total,width,bottom=100*mazout/total+100*gaz/total+100*pac/total+100*elec/total+100*bois/total+100*sol/total+100*cad/total,color='red')
plt.xticks(N, ('2010', '2011','2012', '2013', '2014', '2015'),rotation=90)

plt.yticks(np.arange(0, 101, 10),rotation=90)
plt.ylabel('Part des agents installés par an (en %)')
plt.grid(color='xkcd:silver',alpha=0.3)
```
